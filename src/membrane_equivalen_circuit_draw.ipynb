```python
%matplotlib inline
import matplotlib.pyplot as plt
import schemdraw
import schemdraw.elements as elm
```

```python
fig = plt.figure(figsize=[4, 4])
ax = fig.add_axes([0, 0, 1, 1], xlim=[-2, 5], ylim=[-1, 5])

#ax.set_aspect('equal')
ax.axis('off')

d = schemdraw.Drawing(backend='matplotlib')
d.config(unit=3, fontsize=14)
d += (D1 := elm.Dot()).label(r'$V_\mathrm{e1}$', 'left')
d += (L1 := elm.Line().up(0.5))
d.push()
d += (L1 := elm.Line().right(0.7))
d += (R1 := elm.Resistor().up().label("$r_\mathrm{m}$", "top"))
d += (L1 := elm.Line().left(0.7))
d.pop()
d += (L1 := elm.Line().left(0.8))
d += (R1 := elm.Capacitor().up().label("$c_\mathrm{m}$"))
d += (L1 := elm.Line().right(0.8))
d += (L1 := elm.Line().up(0.5))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{i1}$', 'top')
d += (R1 := elm.Resistor().right().label("$r_\mathrm{i}$"))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{i2}$', 'top')
d += (L1 := elm.Line().down(0.5))
d.push()
d += (L1 := elm.Line().right(0.7))
d += (R1 := elm.Resistor().down().label("$r_\mathrm{m}$", 'top'))
d += (L1 := elm.Line().left(0.7))
d.pop()
d += (L1 := elm.Line().left(0.7))
d += (R1 := elm.Capacitor().down().label("$c_\mathrm{m}$"))
d += (L1 := elm.Line().right(0.7))
d += (L1 := elm.Line().down(0.5))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{e2}$', 'right')

d.draw(ax=ax)
fig.savefig('two_comp_equiv_circuit.pdf', pad_inches=0)
```

```python
fig = plt.figure(figsize=[6, 3])
ax = fig.add_axes([0, 0, 1, 1], xlim=[-4, 11], ylim=[-2.5, 3.9])

h = 1.5

ax.axis('off')

d = schemdraw.Drawing(backend='matplotlib')
d.config(unit=3, fontsize=14)
d += (D1 := elm.Dot()).label(r'$V_\mathrm{e1}$', ofst=(-.4, .0))
d.push()
d += (L1 := elm.Line(color='blue').left(2.5))
d += (L1 := elm.MeterV(color='blue').up(2.5).label("$V_\mathrm{m1}$"))
d += (L1 := elm.Line(color='blue').right(2.5))
d.pop()
d.push()
d += (R1 := elm.Resistor().right().label("$r_\mathrm{e1}$", "bottom"))
d.pop()
d.push()
d += (R1 := elm.Resistor().down(h).label("$r_\mathrm{e0}$", "top"))
d += (D1 := elm.Dot()).label(r"$\vec r_\mathrm{EC}$", 'left')
d += (R1 := elm.SourceSin(color='red').right(2).label("$I_\mathrm{EC}$", "bottom"))
d += (R1 := elm.Ground(color='red'))
d.pop()
d += (L1 := elm.Line().up(0.5))
d.push()
d += (L1 := elm.Line().right(0.7))
d += (R1 := elm.Resistor().up(h).label("$r_\mathrm{m}$", "top"))
d += (L1 := elm.Line().left(0.7))
d.pop()
d += (L1 := elm.Line().left(0.8))
d += (R1 := elm.Capacitor().up(h).label("$c_\mathrm{m}$"))
d += (L1 := elm.Line().right(0.8))
d += (L1 := elm.Line().up(0.5))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{i1}$', 'top')
d += (R1 := elm.Resistor().right().label("$r_\mathrm{i}$"))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{i2}$', 'top')
d.push()
d += (R1 := elm.Resistor().right().label("$r_\mathrm{i}$", "top"))
d.pop()
d += (L1 := elm.Line().down(0.5))
d.push()
d += (L1 := elm.Line().right(0.7))
d += (R1 := elm.Resistor().down(h).label("$r_\mathrm{m}$", 'top'))
d += (L1 := elm.Line().left(0.7))
d.pop()
d += (L1 := elm.Line().left(0.7))
d += (R1 := elm.Capacitor().down(h).label("$c_\mathrm{m}$"))
d += (L1 := elm.Line().right(0.7))
d += (L1 := elm.Line().down(0.5))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{e2}$', 'bottom')

d += (R1 := elm.Resistor().right(3).label("$r_\mathrm{e2}$", "bottom"))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{e3}$', 'bottom')
d.push()
d += (L1 := elm.Line().right(0.8))
d += (L1 := elm.DotDotDot())
d += (L1 := elm.Line().right(0.7))
d += (D1 := elm.Dot()).label(r'$V_{\mathrm{e}n}$', 'right')
d.pop()
d += (L1 := elm.Line().up(0.5))
d.push()
d += (L1 := elm.Line().right(0.7))
d += (R1 := elm.Resistor().up(h).label("$r_\mathrm{m}$", "top"))
d += (L1 := elm.Line().left(0.7))
d.pop()
d += (L1 := elm.Line().left(0.8))
d += (R1 := elm.Capacitor().up(h).label("$c_\mathrm{m}$"))
d += (L1 := elm.Line().right(0.8))
d += (L1 := elm.Line().up(0.5))
d += (D1 := elm.Dot()).label(r'$V_\mathrm{i3}$', 'top')

d += (L1 := elm.Line().right(0.8))
d += (L1 := elm.DotDotDot())
d += (L1 := elm.Line().right(0.7))
d += (D1 := elm.Dot()).label(r'$V_{\mathrm{i}n}$', 'top')
d += (L1 := elm.Line().down(0.5))
d.push()
d += (L1 := elm.Line().right(0.7))
d += (R1 := elm.Resistor().down(h).label("$r_\mathrm{m}$", 'top'))
d += (L1 := elm.Line().left(0.7))
d.pop()
d += (L1 := elm.Line().left(0.7))
d += (R1 := elm.Capacitor().down(h).label("$c_\mathrm{m}$"))
d += (L1 := elm.Line().right(0.7))
d += (L1 := elm.Line().down(0.5))
d += (R1 := elm.Resistor().down(h).label("$r_{\mathrm{e}n}$"))
d += (R1 := elm.Ground())


l = 2.5
d += elm.Coax(xy=(1.2 - l, 2.5), radius=1.25, length=l, leadlen=0.0, color='0.96', zorder=-1).theta(0)
d += elm.Coax(xy=(1.2, 2.5), radius=1.25, length=l, leadlen=0.0, color='0.96', zorder=-1).theta(0)
d += elm.Coax(xy=(1.2 + l, 2.5), radius=1.25, length=l, leadlen=0.0, color='0.96', zorder=-1).theta(0)
d += elm.Coax(xy=(1.2 + 2.5*l, 2.5), radius=1.25, length=l, leadlen=0.0, color='0.96', zorder=-1).theta(0)

d.draw(ax=ax)
fig.savefig('n_comp_equiv_circuit_I_ext.pdf', pad_inches=0)
```
